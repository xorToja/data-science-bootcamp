# importowanie bibliotek

```
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```
img = mpimg.imread('office.jpg')
img # obraz zapisany jest jako obiekt np array
```

```
img.shape # wymiar obrazu to 426 px na 640 px, składa się z 3 kanałów (bo jest kolorowy)
```

```
plt.imshow(img)
plt.axis('Off')
```

```
fig = plt.figure(figsize=(12, 10))
plt.subplot(221)
plt.imshow(mpimg.imread('office.jpg'))
plt.subplot(222)
plt.imshow(mpimg.imread('business.jpg'))
plt.subplot(223)
plt.imshow(mpimg.imread('children.jpg'))
plt.subplot(224)
plt.imshow(mpimg.imread('workplace.jpg'))
```

```
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(220 + idx + 1)
  plt.imshow(mpimg.imread(img))
```

```
fig = plt.figure(figsize=(12, 10))

for idx, img in enumerate(['office.jpg', 'business.jpg', 'children.jpg', 'workplace.jpg']):
  plt.subplot(140 + idx + 1)
  plt.imshow(mpimg.imread(img))
```
