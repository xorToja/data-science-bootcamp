# importowanie biblioteki

```
import matplotlib
matplotlib.__version__
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set() # nakladka na matplotliba
```

# podstawowy obiekt

```
fig = plt.figure()
fig.show()
```

# wykres liniowy

```
_ = plt.plot([0,1], [0,3]) # x: [,] y: [,]
```

```
plt.plot([0,1], [0,3], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```
plt.plot([0,1, 2], [0,3, 2], label='line', color='green')
plt.xlabel('oś x')
plt.ylabel('oś y')
plt.title('Wykres')
_ = plt.legend()
```

```
x = np.arange(-3, 3, 0.1)
x_sin = np.sin(x)
x_cos = np.cos(x)

_ = plt.plot(x, x_sin)
_ = plt.plot(x, x_cos)
```

# Tworzenie wielu wykresów

```
plt.subplot(121) # 1 - liczba wierszy w siatce wykresów, 2 - liczba kolumn w siatce wykresów, 1 - odnosi się do konkretnego wykresu
plt.plot([0,1,2],[3,2,5])
plt.subplot(122)
plt.plot([0,1,2], [3,1,4], 'r--')
```

```
# ręczne tworzenie figure:
fig = plt.figure(figsize = (12, 8))
# bo domyślnie samo tworzyło obiekt typu figure

plt.subplot(131) # czyli tym razem jeden wiersz, ale trzy kolumny wykresów
plt.plot([0,1,2],[3,2,5])
plt.subplot(132)
plt.plot([0,1,2], [3,1,4], 'r--')
plt.subplot(133)
plt.plot([0,1,2], [3,1,4], 'ro')

# w pycharm trzeba użyć polecenia fig.show() żeby zobaczyć wykres
```

# Wykresy słupkowe

```
plt.bar(x=[0,1,2,3], height = [1.2, 3.2, 1.2, 5.2])
```

```
plt.bar(x=['a','b','c','d'], height = [1.2, 3.2, 1.2, 5.2])
```

```
plt.barh(y=['a','b','c','d'], width = [1.2, 3.2, 1.2, 5.2]) # barh - bar horizontal 
```

# wykresy punktowe

```
x1 = np.random.randn(300)
x2 = np.random.randn(300)

plt.scatter(x = x1, y = x2)
```

```
x3 = np.random.randn(300) * 50

plt.scatter(x = x1, y = x2, s = x3, alpha = 0.5) # s - size - rozmiar, można użyć jako 3 wymiaru 
# natomiast alpha to przezroczystość
```
